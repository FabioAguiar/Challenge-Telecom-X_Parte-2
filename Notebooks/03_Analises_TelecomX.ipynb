## 1) Carregamento de artefatos

```python
import matplotlib.pyplot as plt
import os, joblib, numpy as np, pandas as pd
from sklearn.metrics import classification_report, confusion_matrix



MODEL_PATH = "artefatos/modelo_churn_telecomx.pkl"
ARTEFATOS_PATH = "artefatos/artefatos_modelagem.pkl"
TESTSET_PATH = "artefatos/conjunto_teste.pkl"

if not (os.path.exists(MODEL_PATH) and os.path.exists(ARTEFATOS_PATH) and os.path.exists(TESTSET_PATH)):
    raise FileNotFoundError("N√£o encontrei os artefatos esperados. Rode o Notebook 2 at√© a etapa de exporta√ß√£o.")

best_pipe = joblib.load(MODEL_PATH)
artefatos = joblib.load(ARTEFATOS_PATH)
teste = joblib.load(TESTSET_PATH)

X_test, y_test = teste["X_test"], teste["y_test"]

feature_names = artefatos.get("feature_names")
importances_df = artefatos.get("importances_df")
coefs_df = artefatos.get("coefs_df")

print("Origem do melhor modelo:", artefatos.get("origem_melhor_modelo"))
```

## 2) Desempenho no conjunto de teste

```python
import numpy as np
y_pred = best_pipe.predict(X_test)
print(classification_report(y_test, y_pred, digits=3))
```

## 3) Matriz de confus√£o

```python
cm = confusion_matrix(y_test, y_pred)

fig = plt.figure()
plt.imshow(cm, interpolation='nearest')
plt.title('Matriz de Confus√£o ‚Äî Modelo Final')
plt.xlabel('Predito')
plt.ylabel('Real')
for (i, j), val in np.ndenumerate(cm):
    plt.text(j, i, int(val), ha='center', va='center')
plt.show()
```

## 4) Permutation Importance (p√≥s-OHE ‚Äî padr√£o)

```python

from sklearn.inspection import permutation_importance

# Detecta automaticamente o nome do passo de pr√©-processamento
pre_step_name = None
for candidate in ["pre", "preprocess"]:
    if candidate in best_pipe.named_steps:
        pre_step_name = candidate
        break
if pre_step_name is None:
    raise RuntimeError("N√£o encontrei o passo de pr√©-processamento no pipeline ('pre' ou 'preprocess').")

pre = best_pipe.named_steps[pre_step_name]
model = best_pipe.named_steps["model"]

# Transforma X_test para o espa√ßo p√≥s-OHE
X_test_trans = pre.transform(X_test)

# Extrai nomes p√≥s-OHE
ohe = pre.named_transformers_["cat"]
num_names = np.array(pre.transformers_[0][2], dtype=object)  # colunas num√©ricas usadas no CT
cat_cols_ = pre.transformers_[1][2]                           # colunas categ√≥ricas usadas no CT
cat_names = ohe.get_feature_names_out(cat_cols_)
feature_names_perm = np.concatenate([num_names, cat_names])

# Import√¢ncia por permuta√ß√£o do estimador final (dados transformados)
perm = permutation_importance(
    model, X_test_trans, y_test,
    n_repeats=10, random_state=42, scoring="f1", n_jobs=-1
)

perm_df = (
    pd.DataFrame({
        "feature": feature_names_perm,
        "importance_mean": perm.importances_mean,
        "importance_std": perm.importances_std
    })
    .sort_values("importance_mean", ascending=False)
    .reset_index(drop=True)
)
display(perm_df.head(20))
```

## 5) Fatores do modelo (import√¢ncias/coeficientes)

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# Se importances_df/coefs_df n√£o foram salvos, tenta reconstruir rapidamente
if feature_names is None:
    try:
        feature_names = feature_names_perm  # j√° calculados acima
    except NameError:
        feature_names = None

if importances_df is None and isinstance(best_pipe.named_steps['model'], RandomForestClassifier) and feature_names is not None:
    importances = best_pipe.named_steps['model'].feature_importances_
    importances_df = (pd.DataFrame({'feature': feature_names, 'importance': importances})
                      .sort_values('importance', ascending=False)
                      .reset_index(drop=True))

if coefs_df is None and isinstance(best_pipe.named_steps['model'], LogisticRegression) and feature_names is not None:
    coefs = best_pipe.named_steps['model'].coef_.ravel()
    coefs_df = (pd.DataFrame({'feature': feature_names, 'coef': coefs})
                .assign(abs_coef=lambda d: d['coef'].abs())
                .sort_values('abs_coef', ascending=False)
                .reset_index(drop=True))

display(importances_df.head(20) if importances_df is not None else coefs_df.head(20) if coefs_df is not None else "Sem dados de import√¢ncia/coeficientes.")
```

## 6) Gr√°ficos (Top-15 fatores)

```python
def plot_top(df, value_col, label_col="feature", top=15, title="Ranking"):
    if df is None or df.empty:
        print("Nada a plotar:", title)
        return
    top_df = df.nlargest(top, value_col).iloc[::-1]
    fig = plt.figure()
    plt.barh(top_df[label_col], top_df[value_col])
    plt.title(title)
    plt.xlabel(value_col)
    plt.ylabel(label_col)
    plt.tight_layout()
    plt.show()

# Permutation Importance
plot_top(perm_df, "importance_mean", title="Top-15 (Permutation Importance)")

# Espec√≠fico do modelo (se houver)
if 'importances_df' in globals() and importances_df is not None:
    plot_top(importances_df, "importance", title="Top-15 Import√¢ncias (Random Forest)")
if 'coefs_df' in globals() and coefs_df is not None:
    if "abs_coef" not in coefs_df.columns and "coef" in coefs_df.columns:
        coefs_df = coefs_df.assign(abs_coef=lambda d: d["coef"].abs())
    plot_top(coefs_df, "abs_coef", title="Top-15 |Coef| (Logistic Regression)")
```

```python
# %%
# === Tabela executiva de drivers (robusta, com nomes reais de features) ===
import numpy as np
import pandas as pd
import re

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline

# ----------------------------
# Helpers p/ extrair nomes
# ----------------------------
def _safe_name(x):
    try:
        return str(x)
    except Exception:
        return repr(x)

def get_feature_names_from_ct(ct: ColumnTransformer, input_features: list | None = None):
    """
    Retorna a lista de nomes de features GERADAS pelo ColumnTransformer `ct`.
    - Suporta OneHotEncoder, Pipelines, transformers com get_feature_names_out.
    - Respeita remainder='passthrough' (mant√©m colunas originais n√£o transformadas).
    """
    feature_names = []
    original_cols = list(input_features) if input_features is not None else None

    for name, transformer, columns in ct.transformers_:
        if name == 'remainder':
            if ct.remainder == 'passthrough' and original_cols is not None:
                # Colunas usadas explicitamente
                used_cols = []
                for n2, t2, cols2 in ct.transformers_:
                    if n2 != 'remainder' and isinstance(cols2, (list, tuple, np.ndarray, pd.Index)):
                        used_cols.extend([_safe_name(c) for c in cols2])
                # Resto
                remainder_cols = [c for c in original_cols if c not in used_cols]
                feature_names.extend([_safe_name(c) for c in remainder_cols])
            continue

        # Resolve colunas de entrada deste transformer
        if isinstance(columns, slice):
            if original_cols is None:
                cols = []
            else:
                cols = list(original_cols[columns])
        elif callable(columns):
            cols = []  # n√£o resolvemos callable aqui
        else:
            cols = list(columns) if columns is not None else []

        # Se for um Pipeline, pega o √∫ltimo est√°gio (o que gera as sa√≠das)
        trans = transformer
        if isinstance(trans, Pipeline):
            trans = trans.steps[-1][1]

        # OneHotEncoder
        if isinstance(trans, OneHotEncoder):
            try:
                ohe_names = trans.get_feature_names_out([_safe_name(c) for c in cols])
                feature_names.extend(list(ohe_names))
                continue
            except Exception:
                # Fallback simples (menos confi√°vel para drop='first' etc.)
                for base_col, cats in zip(cols, trans.categories_):
                    base = _safe_name(base_col)
                    for cat in cats:
                        feature_names.append(f"{base}_{cat}")
                continue

        # Transformers em geral com get_feature_names_out
        if hasattr(trans, "get_feature_names_out"):
            try:
                names = trans.get_feature_names_out([_safe_name(c) for c in cols]) if len(cols) else trans.get_feature_names_out()
                feature_names.extend(list(names))
                continue
            except Exception:
                pass

        # Sem nomes espec√≠ficos => devolve colunas originais
        if len(cols) > 0:
            feature_names.extend([_safe_name(c) for c in cols])
        else:
            n_out = getattr(trans, 'n_features_out_', 1)
            feature_names.extend([f"{name}__{i}" for i in range(n_out)])

    return feature_names

def get_feature_names_from_pipeline(pipe, X_sample: pd.DataFrame | np.ndarray):
    """
    Vers√£o de alto n√≠vel para Pipeline completo (scikit Pipeline).
    - Procura um passo 'preprocess' do tipo ColumnTransformer; se n√£o achar, tenta detectar.
    - Prefira passar DataFrame (usa X_sample.columns).
    """
    # 1) tenta pelo nome padr√£o
    if hasattr(pipe, "named_steps") and "preprocess" in pipe.named_steps:
        ct = pipe.named_steps["preprocess"]
    else:
        # 2) tenta localizar um CT em qualquer step
        ct = None
        if hasattr(pipe, "steps"):
            for _, step in pipe.steps:
                if isinstance(step, ColumnTransformer):
                    ct = step
                    break
    if ct is None:
        # Sem CT -> tenta names direto
        if hasattr(pipe, "get_feature_names_out"):
            try:
                return list(pipe.get_feature_names_out())
            except Exception:
                pass
        n_cols = X_sample.shape[1] if hasattr(X_sample, "shape") else 0
        return [f"feature_{i}" for i in range(n_cols)]

    input_features = list(X_sample.columns) if hasattr(X_sample, "columns") else None
    return get_feature_names_from_ct(ct, input_features)

# (Opcional) Limpeza de nomes mais amig√°vel
def humanize(name: str):
    name = re.sub(r'(^|\b)(preprocess|num|cat|onehot|scale|imputer|pipeline)\__', '', name)
    name = name.replace('__', '_')
    return name

# ----------------------------
# Permutation Importance + Dire√ß√£o (para LogReg)
# ----------------------------
def _safe_permutation_importance(estimator, X, y, scoring="f1", n_repeats=10, random_state=42):
    try:
        from sklearn.inspection import permutation_importance
        res = permutation_importance(estimator, X, y, scoring=scoring, n_repeats=n_repeats, random_state=random_state)
        return res.importances_mean, res.importances_std
    except Exception as e:
        print("[INFO] N√£o foi poss√≠vel calcular permutation_importance automaticamente:", e)
        return None, None

def _extract_linear_direction(pipe, feature_names):
    # Dire√ß√£o (sinal do coeficiente) se o final for LogisticRegression
    try:
        from sklearn.linear_model import LogisticRegression
        final_est = getattr(pipe, "named_steps", {}).get("model", None) or getattr(pipe, "steps", [[None, None]])[-1][1]
        if isinstance(final_est, LogisticRegression):
            coefs = final_est.coef_.ravel()
            if len(coefs) != len(feature_names):
                m = min(len(coefs), len(feature_names))
                coefs = coefs[:m]
                feature_names = feature_names[:m]
            return dict(zip(feature_names, coefs))
    except Exception:
        pass
    return {}

# ----------------------------
# Montagem da tabela
# ----------------------------
best_pipe_exist = "best_pipe" in globals()
X_test_exist = "X_test" in globals() and "y_test" in globals()

if not (best_pipe_exist and X_test_exist):
    print("[AVISO] Objetos 'best_pipe', 'X_test' e/ou 'y_test' n√£o est√£o no ambiente. "
          "A tabela de drivers ser√° um template vazio para preenchimento manual.")
    tabela_drivers = pd.DataFrame(columns=["Fator", "Import√¢ncia (PI)", "Dire√ß√£o (se aplic√°vel)", "Interpreta√ß√£o"])
else:
    # Garante DataFrame para termos nomes de colunas
    if not hasattr(X_test, "columns"):
        print("[AVISO] X_test n√£o √© DataFrame. Converta para pandas DataFrame com nomes de colunas para nomes de features leg√≠veis.")
    feature_names = get_feature_names_from_pipeline(best_pipe, X_test)

    imp_mean, imp_std = _safe_permutation_importance(best_pipe, X_test, y_test, scoring="f1")
    if imp_mean is None:
        # Tenta recuperar de artefatos pr√©-computados
        tabela = pd.DataFrame(columns=["Fator", "Import√¢ncia (PI)"])
        if "artefatos" in globals():
            pi_df = globals()["artefatos"].get("pi_df")
            if pi_df is not None:
                tmp = pi_df.copy()
                tmp.columns = [c.strip() for c in tmp.columns]
                fcol = next((c for c in tmp.columns if "feature" in c.lower()), None)
                icol = next((c for c in tmp.columns if "mean" in c.lower() or "importance" in c.lower()), None)
                if fcol and icol:
                    tabela = tmp[[fcol, icol]].rename(columns={fcol: "Fator", icol: "Import√¢ncia (PI)"})
    else:
        m = min(len(feature_names), len(imp_mean))
        tabela = pd.DataFrame({
            "Fator": [humanize(f) for f in feature_names[:m]],
            "Import√¢ncia (PI)": imp_mean[:m]
        }).sort_values("Import√¢ncia (PI)", ascending=False).head(15).reset_index(drop=True)

    # Dire√ß√£o (se modelo linear)
    directions = _extract_linear_direction(best_pipe, feature_names[:len(tabela)])
    if directions:
        tabela["Dire√ß√£o (se aplic√°vel)"] = tabela["Fator"].map(
            lambda f: "‚Üë (positivo)" if directions.get(f, 0) > 0 else ("‚Üì (negativo)" if directions.get(f, 0) < 0 else "0 (neutro)")
        )
    else:
        tabela["Dire√ß√£o (se aplic√°vel)"] = ""

    tabela["Interpreta√ß√£o"] = ""  # campo livre para anota√ß√£o executiva
    tabela_drivers = tabela

# Monta dataframe final com interpreta√ß√µes
interpretacoes = {
    "Servico_Telefone": "Ter telefone no pacote reduz a evas√£o (maior engajamento/bundle).",
    "Backup_Online": "Servi√ßo de backup/armazenamento reduz a evas√£o (valor adicional).",
    "Fatura_Diaria": "Cobran√ßa di√°ria aumenta a evas√£o (maior atrito/percep√ß√£o de custo).",
    "Streaming_Filmes": "Pacote de filmes aumenta a evas√£o (perfil promocional/sazonal).",
    "Protecao_Dispositivo": "Seguro/prote√ß√£o de dispositivo aumenta a evas√£o (custo extra/associado a problemas).",
    "Multiplas_Linhas": "M√∫ltiplas linhas aumentam a evas√£o (conta maior, mais sens√≠vel a pre√ßo).",
    "Genero_Female": "Sem efeito consistente sobre churn.",
    "Fatura_Digital": "Fatura digital (paperless) aumenta a evas√£o (perfil mais vol√°til).",
    "Possui_Parceiro": "Ter parceiro(a) reduz a evas√£o (maior estabilidade).",
    "Fatura_Mensal": "Fatura/contrato mensal aumenta a evas√£o (baixa fideliza√ß√£o).",
    "Suporte_Tecnico": "Plano com suporte t√©cnico reduz a evas√£o (melhor resolu√ß√£o de problemas).",
    "Idoso": "Perfil idoso aumenta a evas√£o (poss√≠vel desajuste de plano/uso).",
    "Genero_Male": "Sem efeito consistente sobre churn.",
    "Streaming_TV": "Pacote de TV aumenta a evas√£o (perfil promocional/sazonal).",
    "Possui_Dependentes": "Ter dependentes reduz a evas√£o (planos familiares fixam mais).",
}

tabela_final = tabela_drivers.copy()
tabela_final["Interpreta√ß√£o"] = tabela_final["Fator"].map(interpretacoes)

# Visualizar
tabela_final
```

## Conclus√£o dos Drivers de Churn

A an√°lise de **Permutation Importance** evidencia que os fatores mais cr√≠ticos para evas√£o de clientes s√£o:

- **Servi√ßo de Telefone** e **Backup Online** (efeito protetivo ‚Üí reduzem churn).  
- **Fatura Di√°ria**, **Fatura Mensal** e **Fatura Digital** (efeito de risco ‚Üí aumentam churn, sugerem insatisfa√ß√£o com forma de cobran√ßa).  
- **M√∫ltiplas Linhas** e **Streaming** (TV/Filmes) (efeito de risco ‚Üí perfis promocionais ou mais sens√≠veis a pre√ßo).  
- **Suporte T√©cnico** e **Possuir Parceiro/Dependentes** (efeito protetivo ‚Üí maior fideliza√ß√£o).  

### Implica√ß√µes de Neg√≥cio
- **Foco em planos de fideliza√ß√£o**: reduzir contratos mensais/di√°rios com ofertas anuais.  
- **Cross-sell inteligente**: manter clientes em servi√ßos de valor (telefone, backup), mas avaliar pacotes de streaming para reduzir rotatividade.  
- **Atendimento e suporte**: refor√ßar canais de suporte, pois sua presen√ßa reduz churn.  
- **Segmenta√ß√£o proativa**: priorizar clientes sem dependentes/parceiro(a), com contratos mensais ou digitais, para campanhas de reten√ß√£o.  

> Em resumo, **o churn est√° ligado √† forma de cobran√ßa, sensibilidade a pre√ßo e uso de servi√ßos complementares**, enquanto **servi√ßos b√°sicos e suporte t√©cnico atuam como √¢ncoras de reten√ß√£o**.
